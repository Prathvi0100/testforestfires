## Data Cleaning

```python
## missing values
dataset.isnull().any()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

 The dataset is converted into two sets based on region from 122th index, we 
 can make a new column based on the region

 1 : "Bejai Region Dataset"

 2 : "Sidi-Bel Abbes Region Dataset"

 Add new column with region
 

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df=df.dropna().reset_index(drop=True) ## adding .reset_index(drop=True) makes the index representation proper
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## remove row 122
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fixing space in column names
df.columns=df.columns.str.strip()
df.columns
```

#### Changes the required columns as integer data type

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

#### Changing the other columns to float datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
 df.info()
```
